```python
import shapely.geometry as sg
import matplotlib.pyplot as plt
import descartes
```

```python
def plot_intersection(offset_x=0.0, beta=1.0):
    point_center_A = -0.5 + offset_x
    point_center_B = +0.5 + offset_x
    circle_center_A = -(1-beta)*0.5  + offset_x
    circle_center_B = (1-beta)*0.5  + offset_x 
    radius = beta*0.5
    a = sg.Point(circle_center_A,0).buffer(radius)
    b = sg.Point(circle_center_B,0).buffer(radius)

    middle = a.intersection(b)
    ax.add_patch(descartes.PolygonPatch(middle, fc='black', alpha=0.7))
    plt.scatter([point_center_A,point_center_B], [0, 0 ], color='black', s=100.0)
    plt.text(offset_x-0.6, 1.65, r"$\beta={}$".format(beta), fontsize=20)

# use descartes to create the matplotlib patches
fig = plt.figure(figsize=(10,5))
ax = plt.Axes(fig, [0., 0., 1., 1.])

ax = plt.gca()

plot_intersection(offset_x=0.0, beta=1.0)
plot_intersection(offset_x=1.5, beta=2.0)
plot_intersection(offset_x=3.0, beta=3.0)
plot_intersection(offset_x=4.5, beta=4.0)
plot_intersection(offset_x=6.0, beta=5.0)


plt.axis('equal')
plt.axis('off')
plt.savefig("lunes.pdf", bbox_inches='tight', pad_inches = 0)
```
